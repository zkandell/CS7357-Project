```python
import pandas as pd
```

```python
# Get the data from the CSV file
file = "Datasets/goodreads_data.csv"
rawdata = pd.read_csv(file)

# Drop the Author, Avg_Rating, Num_Ratings, and URL columns
data = rawdata.drop(columns=["Author", "Avg_Rating", "Num_Ratings", "URL"])

# Display the first 5 rows of the data
print(data.head())
```

```python
# The genres column contains lists of genres for each book. First, we need to determine all the unique genres in the dataset.
# Create an empty list to store all the genres
genres = []

# Iterate over the rows of the data
for index, row in data.iterrows():
    # Get the genres of the book
    book_genres = row["Genres"]
    # Remove the square brackets, split the genres, and remove all quotes
    book_genres = book_genres[1:-1].split(", ")
    book_genres = [genre.replace("'", "") for genre in book_genres]
    # Add the genres to the list
    genres.extend(book_genres)

# Get the unique genres
unique_genres = list(set(genres))

# Print the unique genres
print(unique_genres)

# Get the number of unique genres
num_genres = len(unique_genres)
print(num_genres)
```

```python
# Add a column for each genre to the data
# Start by copying the data
data_with_genres = data.copy()

# Each genre will have a value of 0 or 1, depending on whether the book belongs to that genre
# One-hot encode the genres
for genre in unique_genres:
    data_with_genres[genre] = data_with_genres["Genres"].str.contains(genre).astype(int)

# Drop the Genres column
data_with_genres = data_with_genres.drop(columns=["Genres"])

# Display the first 5 rows of the data with genres
print(data_with_genres.head())
```

```python
# Identify the number of books in each genre
# Create an empty dictionary to store the number of books in each genre
genre_counts = {}

# Iterate over the unique genres
for genre in unique_genres:
    # Count the number of books in the genre
    count = data_with_genres[genre].sum()
    # Add the count to the dictionary
    genre_counts[genre] = count

# Print the number of books in each genre
print(genre_counts)

# Print the genre counts in descending order
sorted_genre_counts = sorted(genre_counts.items(), key=lambda x: x[1], reverse=True)
print(sorted_genre_counts)

# Get the number of genres with fewer than 100 books
num_few_books = sum(count < 100 for count in genre_counts.values())
print(num_few_books)

# Drop genres with fewer than 100 books
for genre, count in genre_counts.items():
    if count < 100:
        data_with_genres = data_with_genres.drop(columns=[genre])

# Remove the '' genre
data_with_genres = data_with_genres.drop(columns=[""])

# Get the top 25 genres by count, not including the '' genre
top_genres = [genre for genre, count in sorted_genre_counts if genre != ""][:25]
```

```python
print(top_genres)
```

```python
# print(data_with_genres.head())
# Create a new DataFrame with only the top 25 genres
data_top_genres = data_with_genres[["Book", "Description"] + top_genres]
print(data_top_genres.shape)
```

```python
# Remove any books that do not belong to any genre
data_top_genres = data_top_genres[data_top_genres.iloc[:, 4:].sum(axis=1) > 0]
print(data_top_genres.head())
print(data_top_genres.shape)
```

```python
# Save the data with genres to a new CSV file
data_top_genres.to_csv("Datasets/goodreads_data_onehot_genres.csv", index=False)
```
